```python
import numpy as np
from cvxopt import matrix, solvers
import cvxpy as cp

# Step 1: Initialize MDP with arbitrary size and random rewards
def initialize_mdp(S, A, R_min=0, R_max=10):
    gamma = 0.9  # Discount factor

    # Randomly generate transition probabilities P(s'|s,a)
    P = np.random.rand(S, S, A)
    for s in range(S):
        for a in range(A):
            P[s, :, a] /= np.sum(P[s, :, a])  # Normalize to make valid probabilities

    # Randomly generate state-dependent rewards (integers between R_min and R_max)
    R = np.random.randint(R_min, R_max + 1, size=S)

    return S, A, P, R, gamma

# Step 2: Solve for Optimal Policy using the known MDP
def value_iteration(P, R, gamma):
    S, A = P.shape[0], P.shape[2]
    V = np.zeros(S)
    R_sa = np.tile(R, (A, 1)).T  # Shape (S, A)
    tol = 1e-6  # Tolerance for convergence
    while True:
        Q = np.zeros((S, A))
        for a in range(A):
            Q[:, a] = R_sa[:, a] + gamma * P[:, :, a] @ V
        V_new = np.max(Q, axis=1)
        if np.max(np.abs(V_new - V)) < tol:
            break
        V = V_new
    pi = np.argmax(Q, axis=1)
    return pi, V

# Step 3: New IRL method maximizing Q differences
def irl_q_difference(S, A, P, pi, gamma, Rmin=0, Rmax=10, epsilon=1e-3, lambda_reg=0.1):
    # Variables
    R = cp.Variable(S)
    V = cp.Variable(S)
    delta = cp.Variable((S, A))

    # Constraints
    constraints = []

    # Bellman equations for policy pi
    P_pi = np.array([P[s, :, pi[s]] for s in range(S)])  # Shape: (S, S)
    constraints.append(V == R + gamma * P_pi @ V)

    # Reward bounds
    constraints.append(R >= Rmin)
    constraints.append(R <= Rmax)

    # Optimality conditions and slack variables for Q differences
    q_diff = []
    for s in range(S):
        for a in range(A):
            Q_sa = R[s] + gamma * P[s, :, a] @ V
            if a != pi[s]:
                constraints.append(delta[s, a] == V[s] - Q_sa)
                constraints.append(delta[s, a] >= epsilon)
                q_diff.append(delta[s, a])
            else:
                constraints.append(delta[s, a] == 0)

    # Objective function: Maximize total Q difference minus regularization
    objective = cp.Maximize(cp.sum(q_diff) - lambda_reg * cp.norm(R, 1))

    # Solve the optimization problem
    prob = cp.Problem(objective, constraints)
    prob.solve()

    # Retrieve the optimal reward function
    return R.value

# Step 4: LP-based IRL using CVXOPT
def lp_irl_cvxopt(P, pi, gamma, R_max=10, l1=0.1):
    S, _, A = P.shape
    c = np.hstack([np.zeros(S), -np.ones(S), l1 * np.ones(S)])
    num_constraints = 2 * S * (A - 1) + 4 * S
    G = np.zeros((num_constraints, 3 * S))
    h = np.zeros(num_constraints)
    cnt = 0
    for s in range(S):
        a_opt = pi[s]
        P_pi = np.array([P[s_prime, :, a_opt] for s_prime in range(S)])
        inv_matrix = np.linalg.inv(np.eye(S) - gamma * P_pi)
        for a in range(A):
            if a != a_opt:
                diff = (P[s, :, a_opt] - P[s, :, a]) @ inv_matrix
                G[cnt, :S] = -diff
                G[cnt + S * (A - 1), :S] = -diff
                G[cnt + S * (A - 1), S + s] = 1
                cnt += 1
    # Reward bounds and regularization constraints
    for s in range(S):
        idx = 2 * S * (A - 1) + s
        G[idx, s] = 1  # R <= R_max
        h[idx] = R_max
        G[idx + S, s] = -1  # -R <= 0
        h[idx + S] = 0
        G[idx + 2 * S, s] = 1  # R - u <= 0
        G[idx + 2 * S, 2 * S + s] = -1
        h[idx + 2 * S] = 0
        G[idx + 3 * S, s] = -1  # -R - u <= 0
        G[idx + 3 * S, 2 * S + s] = -1
        h[idx + 3 * S] = 0
    # Solve LP
    sol = solvers.lp(matrix(c), matrix(G), matrix(h))
    R_recovered = np.array(sol['x'][:S]).flatten()
    R_recovered = (R_recovered - R_recovered.min()) / (R_recovered.max() - R_recovered.min()) * R_max
    return R_recovered

# Main code
if __name__ == "__main__":
    # Set arbitrary sizes for S and A
    S = 10  # Number of states
    A = 5  # Number of actions

    # Initialize MDP
    S, A, P, R_true, gamma = initialize_mdp(S, A)
    print("Original Reward Function R:")
    print(R_true)

    # Compute optimal policy
    pi_opt, V_opt = value_iteration(P, R_true, gamma)
    print("\nOptimal Policy pi:")
    print(pi_opt)

    # Run LP-based IRL using CVXOPT
    R_recovered_cvxopt = lp_irl_cvxopt(P, pi_opt, gamma, l1=0.1)
    print("\nRecovered Reward Function from CVXOPT IRL:")
    print(np.round(R_recovered_cvxopt, 2))

    # Run the new IRL method maximizing Q differences
    R_recovered_q_diff = irl_q_difference(S, A, P, pi_opt, gamma, epsilon=0.1, lambda_reg=0.1)
    print("\nRecovered Reward Function from Q-difference IRL:")
    print(np.round(R_recovered_q_diff, 2))
```
